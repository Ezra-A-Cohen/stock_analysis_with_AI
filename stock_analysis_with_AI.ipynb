```python
#Market Cap: The value of a company that is traded on the stock market, 
#calculated by multiplying the total number of shares by the present share price

start = "2014-01-01"
end = '2019-1-01'

tcs = yf.download('TCS',start,end,progress=False)#TCS
infy = yf.download('INFY',start,end,progress=False)#Infosys
wipro = yf.download('WIPRO.NS',start,end,progress=False)#WIPRO
blk = yf.download('BLK',start,end,progress=False)#BlackRock

tcs["Market Cap"]=tcs["Volume"]*tcs["Open"]
infy["Market Cap"]=infy["Volume"]*infy["Open"]
wipro["Market Cap"]=wipro["Volume"]* wipro["Open"]
blk["Market Cap"]=blk["Volume"]*blk["Open"]

tcs["Market Cap"].plot(label ="TCS")
infy["Market Cap"].plot(label ="INFOSYS")
wipro["Market Cap"].plot(label = "WIPRO")
blk["Market Cap"].plot(label ="BlackRock")
plt.legend()
plt.title("Market Cap of Famous Companies")
plt.show()
```

```python
"""
Rolling Mean n: the unweighted mean/average of the last n values.
RM 50 
1. Na N 
50   (1... 50)/50 50 
51  (2.. 51) / 50
52  (3.. 52) / 50
53.  (4...53)/50 
"""
AMZN["RM50"]=AMZN["Open"].rolling(50).mean()
print(AMZN["RM50"])
```

```python
AMZN["RM200"]=AMZN["Open"].rolling(200).mean()
AMZN["RM50"].plot(label ="Rolline mean for Last 50 days")
AMZN["RM200"].plot(label ="Rolline mean for Last 200 days")
AMZN["Open"].plot(label= "Open prices")
plt.legend()
plt.title("Rolling mean predictions")
plt.show()
```

```python
msft["RM3"]=msft["Open"].rolling(3).mean()
msft["RM5"]=msft["Open"].rolling(5).mean()
msft["RM7"]=msft["Open"].rolling(7).mean()

msft["RM3"].plot(label ="Rolline mean for Last 3 days")
msft["RM5"].plot(label ="Rolline mean for Last 5 days")
msft["RM7"].plot(label ="Rolline mean for Last 7 days")
msft["Open"].plot(label= "Open prices")
plt.legend()
plt.title("Rolling mean predictions")
plt.show()
""" 
RM7 because it takes more datapoints before averaging and so is less likely to be overly impacted by an outlier
"""
```

```python
"""
Rolling Mean n: the unweighted mean/average of the last n values.
We assume that the predicted stock price of a particular day is the average of the stock prices of the last 50 days.
Day 1: Avg(-50 to -1 of previous dataset)
Day 2: Avg(-51 to -1 and Day 1 predicted price)
...

"""

rm=50
def predictions_msft(actual_list,rm):
  """ Actual list is the actual Open prices for Microsoft.
  Returns the predicted Open Prices for the next 50 days for Microsoft using the rolling means method
  
  """
  predicted_list=[]
  for i in range(len(actual_list)):
    x=i-1
    y=x-rm
    List=[]
    while(y<x):
      List.append(msft["Open"][y])
      y=y+1
    predicted_list.append(sum(List)/len(List))
  return [predicted_list]
  
print(predictions_msft(msft["Open"],rm))

#There is a flaw in the method in that the first 50 values are all impacted by the data points at the end which shouldn't be how it's done because that doesn't go in chronological order, but when the instructions for this part of the code were given they told us what they wanted the end result to be and gave an example of what it should be doing in which they mentioned doing this for the first couple of days
```

```python
google = yf.download('GOOGL',start,end,progress=False)#Google
facebook = yf.download('FB',start,end,progress=False)#Facebook
amazon = yf.download('AMZN',start,end,progress=False)#Amazon

""" 
Percentage CHange = (Day2 - Day1)/ Day1 = Day2/Day1 - 1
The shift(1) method for a pandas series helps shift values in a column up or down.
"""

google["change"] = (google["Close"]/google["Close"].shift(1) ) - 1
facebook["change"] = (facebook["Close"]/facebook["Close"].shift(1) ) - 1
amazon["change"] = (amazon["Close"]/amazon["Close"].shift(1) ) - 1
facebook["change"].hist(label ="facebook Stock prices")
amazon["change"].hist(label ="amazon Stock prices")
google["change"].hist(label ="Google Stock prices")
plt.legend()
plt.title("Percentage Change in Closing Prices")
plt.show()
```
